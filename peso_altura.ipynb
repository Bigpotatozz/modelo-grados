```python
#Se lee el csv
import pandas
data = pandas.read_csv('peso-altura.csv')
```

```python
#Se separan los datos en arreglos diferentes
X = data['altura']
y = data['peso']
```

```python
#Se parsean a un arreglo de dos dimensiones
reshape_x = X.values.reshape(-1, 1)
reshape_y = y.values.reshape(-1, 1)
```

```python
#Se crea el modelo de regresion lineal y se entrena
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(reshape_x, reshape_y)
```

```python
#Se comprueba la calificacion del entrenamiento del modelo
model.score(reshape_x, reshape_y)
```

```python
#Se crea la grafica de regresion lineal como apoyo
import seaborn
seaborn.scatterplot(x="altura", y="peso", data=data)
```

```python
#Se prueba el modelo con una prediccion de una altura no conocida
model.predict([[1.72]])
```
